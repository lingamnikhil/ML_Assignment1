```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# from google.colab import drive
# drive.mount('/content/drive')
# import pandas as pd
file_path = 'D3.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
def compute_loss(X, y, theta):

    predictions = X.dot(theta) # multiply element by element then sum up
    errors = np.subtract(predictions, y)
    sqrErrors = np.square(errors)
    m = len(y)  # Number of training examples
    J = 1 / (2 * m) * np.sum(sqrErrors) # loss function
    return J
```

```python
def gradient_descent(X, y, theta, alpha, iterations):


    m = len(y)  # Number of training examples
    loss_history = np.zeros(iterations)

    # for loop to iterate for theta values to minimize loss (loss)
    for i in range(iterations):
        predictions = X.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha / m) * X.transpose().dot(errors)
        theta -= sum_delta
        loss_history[i] = compute_loss(X, y, theta)

    return theta, loss_history
```

```python
theta = [0., 0.]
iterations = 1500;
alpha = 0.05;
```

```python
def plot_model(X, y, theta, x_label, y_label, title ):
    # Assuming that X, y, and theta are already defined
    # Also assuming that X has two columns: a feature column and a column of ones

    # New figure
    plt.figure()

    # Scatter plot for the training data
    plt.scatter(X[:, 1], y, color='red', marker='+', label='Training Data')

    # Line plot for the linear regression model
    plt.plot(X[:, 1], X.dot(theta), color='green', label='Linear Regression')

    # Plot customizations
    plt.rcParams["figure.figsize"] = (6, 8)
    plt.grid(True)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend()

    # Show the plot
    plt.show()
```

```python
def plot_loss(loss_history, title):
    iterations = len(loss_history)
    plt.figure() # New figure
    plt.plot(range(1, iterations + 1), loss_history, color='blue')
    plt.rcParams["figure.figsize"] = (6, 8)
    plt.grid(True)

    plt.xlabel('Number of iterations')
    plt.ylabel('loss (J)')
    plt.title(title)

    # Show the plot
    plt.show()
```

```python
vars = ['X1', 'X2', 'X3'] # list of X strings
alpha_dict = {'X1':0.01, 'X2':0.01, 'X3':0.01} # can explore by changing alpha value

for var in vars:
    X = df[var].to_numpy()
    y = df['Y'].to_numpy()
    m = len(y)  # Number of training examples

    X_0 = np.ones((m, 1))
    X_1 = X.reshape(m, 1)
    X = np.hstack((X_0, X_1))

    theta = [0., 0.]
    iterations = 1500
    alpha = alpha_dict[var]

    theta, loss_history = gradient_descent(X, y, theta, alpha, iterations)
    print('Training for ' + var)
    print('Final value of theta =', theta)
    print('loss_history =', loss_history)
    plot_model(X, y, theta, var, 'Y', 'Regression Model for ' + var)
    plot_loss(loss_history, 'Convergence of Gradient Descent Loss for ' + var)
```
